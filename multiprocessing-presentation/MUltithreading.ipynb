# Multithreading

```python

import time 

def cal_square(num):
    print("Calculate square numbers")
    for n in num:
        time.sleep(1)
        print ("square : {}".format(n*n))
        
def cal_cube(num):
    print("calculate cube of numbers")
    for n in num:
        time.sleep(1)
        print("cube : {}".format(n*n*n))
        
arr = [2,4,6,8]

t = time.time() #current time

cal_square(arr)
cal_cube(arr)

print("In time : ",time.time()-t) #time required to excecuation

print("Completed")
```

```python
import time
import threading

def calc_square(numbers):
    print("calculate square numbers")
    for n in numbers:
        time.sleep(1)
        print('square:',n*n)

def calc_cube(numbers):
    print("calculate cube of numbers")
    for n in numbers:
        time.sleep(1)
        print('cube:',n*n*n)

arr = [2,3,8,9]

t = time.time()  #current time
                    
t1= threading.Thread(target=calc_square, args=(arr,)) #creat th1
t2= threading.Thread(target=calc_cube, args=(arr,))   #creat th2

t1.start()        #start the thread
t2.start()

t1.join()         #wait until thread is completed
t2.join()

print("done in : ",time.time()-t)         #time required to excecuation

print("Hah... I am done with all my work now!")
```
